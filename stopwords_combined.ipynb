```python
import numpy as np
import pandas as pd
import os
import nltk
import re
from nltk.tokenize import word_tokenize
nltk.download('punkt')
```

```python
import chardet

def detect_encoding(file_path):
    with open(file_path, 'rb') as file:
        result = chardet.detect(file.read())
    return result['encoding']
```

```python
print(re.search("StopWords",'/content/StopWords_Names.txt').start())
```

```python
path="/content/"
dir_list = os.listdir(path)
print(dir_list)
newList=[]
for filename in dir_list:
  if "StopWords" in filename:
    newList.append(filename)

print(newList)

stops=''
for file in newList:
  file_encoding = detect_encoding(file)
  with open(file, 'r+', encoding=file_encoding) as content:
    data1=content.read()
    data1=data1.upper()
    data1=data1.strip(' ')
    stops+=data1


file_encoding_n = detect_encoding("/content/negative-words.txt")
with open("/content/negative-words.txt", 'r+', encoding=file_encoding_n) as content_n:
    negatives=content_n.read()
    negatives=negatives.upper()
negatives=negatives.split('\n')
print(type(negatives))


file_encoding_p = detect_encoding("/content/positive-words.txt")
with open("/content/positive-words.txt", 'r+', encoding=file_encoding_p) as content_p:
    positives=content_p.read()
    positives=positives.upper()
positives=positives.split('\n')
```

```python
df=pd.read_excel('/content/updated_input.xlsx')
df.head()
```

```python
punctuations = ["'", ".", ",", "?", "/","(", ")", "#", "@", "!", "'", ";", "-",":","’",'”',"$",'“',"%"]
```

```python
noStopText=[]
for i in range(len(df)):
  article = df.iloc[i]['text']
  if (type(article)==str):
    article=article.upper()
    tokens = word_tokenize(article)
    new = []
    for word in tokens:
      if word not in stops and word not in punctuations and word != None:
        new.append(word)
  else:
    new=[]
  noStopText.append(new)
df["Clean Text"]=noStopText
```
